```python
#加载所需的库
import numpy as np
import pandas as pd
```

```python
# 载入data文件中的:train-left-up.csv
df = pd.read_csv('../titanic/train-left-up.csv')
df
```

```python
text_left_up = pd.read_csv("../titanic/train-left-up.csv")
text_left_down = pd.read_csv("../titanic/train-left-down.csv")
text_right_up = pd.read_csv("../titanic/train-right-up.csv")
text_right_down = pd.read_csv("../titanic/train-right-down.csv")
```

```python
#使用concat方法：将数据train-left-up.csv和train-right-up.csv横向合并为一张表，并保存这张表为result_up
result_up = pd.concat([text_left_up,text_right_up],axis=1)
result_up
```

```python
#使用concat方法：将train-left-down和train-right-down横向合并为一张表，并保存这张表为result_down。
result_down = pd.concat([text_left_down,text_right_down],axis=1)
result_down
```

```python
#然后将上边的result_up和result_down纵向合并为result。
result = pd.concat([result_up,result_down])
result
```

```python
#使用DataFrame自带的方法join方法和append：完成任务二和任务三的任务
#join任务二
join_up = text_left_up.join(text_right_up)
join_up
```

```python
join_down = text_left_down.join(text_right_down)
join_down
```

```python
#join任务三
join = join_up._append(join_down)
join
```

```python
merge_up = pd.merge(text_left_up, text_right_up,left_index=True,right_index=True)
merge_up
```

```python
merge_down = pd.merge(text_left_down, text_right_down,left_index=True,right_index=True)
merge_down
```

```python
merge = merge_up._append(merge_down)
merge
```

```python
#完成的数据保存为result.csv
result.to_csv('../titanic/result.csv')
```
