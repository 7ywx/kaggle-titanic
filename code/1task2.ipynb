```python
#加载文件"test_1.csv"，然后对比"train.csv"，看看有哪些多出的列，然后将多出的列删除
df1=pd.read_csv('../titanic/test_1.csv')
# df1.pop('a')
# del df1['a']
# df1.drop(['a'],axis=1,inplace=True) axis={0:索引, 1:列}, inplae={true=真删, false=返回副本}
df1
```

```python
df1.drop(['PassengerId','Name','Age','Ticket'],axis=1)
```

```python
#显示年龄在10岁以下的乘客信息
df1[df1['Age']<10]
```

```python
#将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage
midage=df1[(df1['Age']>10) & (df1['Age']<50)]
midage1=midage.reset_index(drop=True)
midage.to_csv("../titanic/midage.csv")
midage1.to_csv("../titanic/midage1.csv")
midage.head()
```

```python
#将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来
midage1.loc[[100],['Pclass','Sex']]
```

```python
#使用loc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来
midage1.loc[[100,105,108],["Pclass","Name","Sex"]]
```

```python
#使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来
midage1.iloc[[100,105,108],[3,4,5]]
```
